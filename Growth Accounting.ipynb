```python
# Growth Accounting

import numpy as np
import pandas as pd

pd.options.mode.chained_assignment = None

df = pd.read_csv('growth_accounting.csv')
df_1 = df.loc[(df['year'] >= 1960) & (df['year'] <= 2000)]
df_2 = df_1.loc[df_1['country'].isin(['Australia', 'Austria', 'Belgium', 'Germany', 'France', 'United Kingdom', 'United States'])]


df_2['GDP_per_worker'] = (df_2['cgdpe'] + df_2['cgdpo']) / df_2['pop']
df_2['capital_per_worker'] = df_2['cn'] / df_2['pop']

df_2['gy'] = df_2['GDP_per_worker'].pct_change(periods=1) * 100
df_2['gk'] = df_2['capital_per_worker'].pct_change(periods=1) * 100
df_2['gz'] = df_2['ctfp'].pct_change(periods=1) * 100
df_2['alpha'] = (df_2['gy'] - df_2['gz']) / df_2['gk']

df_3 = df_2[['country', 'year', 'gy', 'gz', 'gk', 'alpha']].copy()

df_3.loc[df_3['year'] == 1960, 'gy'] = 'NaN'
df_3.loc[df_3['year'] == 1960, 'gz'] = 'NaN'
df_3.loc[df_3['year'] == 1960, 'gk'] = 'NaN'
df_3.loc[df_3['year'] == 1960, 'alpha'] = 'NaN'

df_3.head(45)
```
